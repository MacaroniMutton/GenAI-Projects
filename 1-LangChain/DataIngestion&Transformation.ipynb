```python
from langchain_community.document_loaders import WebBaseLoader
import bs4

loader = WebBaseLoader(web_paths=("https://jalammar.github.io/illustrated-transformer/",), bs_kwargs=dict(parse_only=bs4.SoupStrainer(class_=("post",))))
docs = loader.load()
docs
```

```python
from langchain_community.document_loaders import ArxivLoader

loader = ArxivLoader(query="1706.03762")
docs = loader.load()
docs
```

```python
from langchain_community.document_loaders import WikipediaLoader

loader = WikipediaLoader(query="HUNTER X HUNTER", load_max_docs=2)
docs = loader.load()
docs
```

```python
from langchain_text_splitters import RecursiveCharacterTextSplitter

text_splitter = RecursiveCharacterTextSplitter(chunk_size=500, chunk_overlap=50)
final_docs = text_splitter.split_documents(docs)
final_docs
```

```python
from langchain_text_splitters import HTMLHeaderTextSplitter
from bs4 import BeautifulSoup
import requests

url = "https://www.techtarget.com/searchenterpriseai/definition/generative-AI"
response = requests.get(url)
soup = BeautifulSoup(response.text, 'html.parser')
soup
```

```python
url = "https://www.techtarget.com/searchenterpriseai/definition/generative-AI"

headers_to_split_on = [
    ("h1", "Header 1"),
    ("h2", "Header 2"),
    ("h3", "Header 3"),
]

text_splitter = HTMLHeaderTextSplitter(headers_to_split_on=headers_to_split_on)
text_splitter.split_text_from_url(url)
```

```python
import json
import requests

response = requests.get("https://api.smith.langchain.com/openapi.json")
json_data = response.json()
json_data
```

```python
from langchain_text_splitters import RecursiveJsonSplitter

text_splitter = RecursiveJsonSplitter(max_chunk_size=300)
json_chunks = text_splitter.split_json(json_data)
```
